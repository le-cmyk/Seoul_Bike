## Visualisation de l'evolution du nombre de vélo loué sur toute la période

```python
def visu_fichier():
    import plotly.graph_objects as px
    dff=df.copy()
    dff["Datetime"]=pd.to_datetime(dff['Date'].astype(str)+"-"+ dff['Hour'].astype(str), format='%d/%m/%Y-%H')
    fig= px.Figure(data=[px.Scatter(
        x=dff["Datetime"],
        y=dff["Rented Bike Count"],
        mode='lines')
    ])
    fig.update_layout(
        title="<b>Evolution</b> of rented Bike in Seoul",
        xaxis_title="Time",
        yaxis_title="Number of rented bike",
        legend_title="Legende",
        font=dict(
            family="Calibri",
            size=15,
            color="Black"
        )
    )
    fig.update_layout(
        xaxis=dict(
            rangeselector=dict(
                buttons=list([
                    dict(count=1,
                         label="1j",
                         step="day",
                         stepmode="backward"),
                    dict(count=2,
                         label="2j",
                         step="day",
                         stepmode="backward"),
                    dict(count=5,
                         label="5j",
                         step="day",
                         stepmode="backward"),                    
                    dict(count=30,
                         label="30j",
                         step="day",
                         stepmode="backward"),
                    dict(step="all")
                ])
            ),
            rangeslider=dict(
                visible=True
            ),
            type="date"
        )
        
    )
    fig.show()
```

# Now let's find the best model

```python
df=pd.read_csv(r"SeoulBikeData.csv", encoding="ISO_8859-1")
```

```python
df["Datetime"]=pd.to_datetime(df['Date'].astype(str)+"-"+ df['Hour'].astype(str), format='%d/%m/%Y-%H')
df['Day'] = pd.DataFrame([date.weekday() for date in df['Datetime']]).astype(str)
del df["Datetime"]
```

```python

target=df["Rented Bike Count"].copy()
data = df.drop(columns=["Rented Bike Count"])

from sklearn.model_selection import train_test_split

data_train, data_test, target_train, target_test = train_test_split(
    data, target, random_state=42)
```

```python
from sklearn.compose import make_column_selector as selector

categorical_columns_selector = selector(dtype_include=object)
categorical_columns = categorical_columns_selector(data)

numerical_features = data.select_dtypes(include='number').columns.tolist()
```

```python
from sklearn.preprocessing import OrdinalEncoder,MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

categorical_preprocessor = OrdinalEncoder(handle_unknown="use_encoded_value",
                                          unknown_value=-1)

numeric_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='mean')),
    ('scale', MinMaxScaler())
])

categorical_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('OrdinalEncoder', OrdinalEncoder(handle_unknown="use_encoded_value",
                                          unknown_value=-1))
])
```

```python
numeric_pipeline.fit_transform(data_train.select_dtypes(include='number'))
```

```python

from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(transformers=[
    ('number', numeric_pipeline, numerical_features),
    ('category', categorical_pipeline, categorical_columns)],
                                 remainder='passthrough', sparse_threshold=0)
```

```python
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.pipeline import Pipeline

model = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier",
     HistGradientBoostingRegressor(random_state=42, max_leaf_nodes=4))])
model
```

```python
%%time
from sklearn.model_selection import GridSearchCV

param_grid = {
    'classifier__learning_rate': (0.01, 0.1, 1, 10),
    'classifier__max_leaf_nodes': (58,60,62,64,66,68,70,73,75)}
model_grid_search = GridSearchCV(model, param_grid=param_grid,
                                 n_jobs=2, cv=2,scoring='neg_mean_absolute_error')
model_grid_search.fit(data_train, target_train)
```

```python
model_grid_search.best_estimator_ 
```

```python
f"the best score while tunning the model was : " f"{model_grid_search.best_score_}"
```

```python
accuracy = model_grid_search.score(data_test, target_test)
print(
    f"The test accuracy score of the grid-searched pipeline is: "
    f"{accuracy:.2f}"
)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
preds = model_grid_search.predict(data_test)
mean_absolute_error(target_test, preds)
```

```python
def mean_absolute_error_model(param_grid,model_debut):
    model = Pipeline([
        ("preprocessor", preprocessor),
        ("classifier",
         model_debut)])
    model_grid_search = GridSearchCV(model, param_grid=param_grid,
                                 n_jobs=2, cv=2,scoring='neg_mean_absolute_error')
    model_grid_search.fit(data_train, target_train)
    print(model_grid_search.best_estimator_ )
    print(f"the best score while tunning the model was : " f"{model_grid_search.best_score_}")
    accuracy = model_grid_search.score(data_test, target_test)
    print(
        f"The test accuracy score of the grid-searched pipeline is: "
        f"{accuracy:.2f}"
    )
    preds = model_grid_search.predict(data_test)
    return (mean_absolute_error(target_test, preds),model_debut)
```

```python
from sklearn.svm import SVC
param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001],'kernel': ['rbf', 'poly', 'sigmoid']}
mean_absolute_error_model(param_grid,SVC())
```
